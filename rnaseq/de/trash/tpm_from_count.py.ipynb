## Convert from featureCounts Count table to TPM
The python code has more accurate result compared with R code (sum back to 1 Million)

```python
import pandas as pd
```

```python
df = pd.read_csv("counts.gene_id.txt", sep="\t", header=0, skiprows=1, index_col=0)
df.head(3)
```

```python
## Fix columns
df.columns = [x.replace(".sort.bam", "") for x in df.columns]
df.columns = [x.replace("../bam/", "") for x in df.columns]
df.columns
```

```python
counts = df.iloc[:, 5:]
print(counts.iloc[0:3, 0:3])
print(counts.shape)
```

```python
length = df['Length']
print(length.head(3))
print(length.shape)
```

```python
RPK = counts.divide(length/1000, axis='rows')
RPK.head()
```

```python
sum_RPK = RPK.sum()
print(sum_RPK[0:3])
```

```python
TPM = RPK/sum_RPK * 1000000
TPM.head()
```

```python
TPM.sum()
```

```python
TPM.insert(0, "Length", length)
```

```python
TPM.head()
```

```python
TPM.to_csv('tpm.csv')
```
